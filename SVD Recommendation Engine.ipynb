```python
ratings= df_ratings.pivot(index= 'userId', columns= 'movieId', values= 'rating').fillna(0)
ratings.head()
```

```python
from surprise import Reader, Dataset, SVD
from surprise.model_selection import cross_validate

reader= Reader()

svd= SVD()

data= Dataset.load_from_df(df_ratings[['userId', 'movieId', 'rating']], reader)

cross_validate(svd, data, measures= ['RMSE', 'MAE'], cv= 3, verbose= True)
```

```python
rating_1= df_ratings[(df_ratings['userId']== 1) & (df_ratings['rating']==5)]
rating_1= rating_1.set_index('movieId')
rating_1= rating_1.join(df_movies)['title']
rating_1.head()
```

```python
user_1= df_movies.copy()

user_1= user_1.reset_index()

data= Dataset.load_from_df(df_ratings[['userId', 'movieId', 'rating']], reader)

trainset= data.build_full_trainset()
svd.fit(trainset)

user_1['Estimate_score']= user_1['movieId'].apply(lambda x: svd.predict(1,x).est)
user_1= user_1.drop(['movieId', 'genres', 'index'], axis= 1)

user_1= user_1.sort_values('Estimate_score', ascending= False)
```
